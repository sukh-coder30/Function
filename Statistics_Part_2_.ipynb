1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import scipy.stats as stats
import numpy as np

# Function to perform a one-sample Z-test
def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Interpret the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean: {population_mean:.2f}")
    print(f"Z-Statistic: {z_statistic:.2f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. The sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. The sample mean is not significantly different from the population mean.")

# Example usage
if __name__ == "__main__":
    # Input data
    sample_data = [100, 102, 98, 101, 99, 103, 97, 105, 100, 102]  # Example sample data
    population_mean = 100  # Known population mean
    population_std = 5     # Known population standard deviation
    alpha = 0.05           # Significance level

    # Perform the Z-test
    z_test(sample_data, population_mean, population_std, alpha)
```

4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Function to perform a two-tailed Z-test
def two_tailed_z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Calculate sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the critical values for the decision region
    z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed test
    lower_bound = -z_critical
    upper_bound = z_critical

    # Calculate the p-value
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Output results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean: {population_mean:.2f}")
    print(f"Z-Statistic: {z_statistic:.2f}")
    print(f"P-Value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. The sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. The sample mean is not significantly different from the population mean.")

    # Visualization
    visualize_decision_region(z_statistic, lower_bound, upper_bound)

# Function to visualize the decision region
def visualize_decision_region(z_statistic, lower_bound, upper_bound):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Shade the rejection regions
    plt.fill_between(x, y, 0, where=(x <= lower_bound) | (x >= upper_bound), color='red', alpha=0.5, label="Rejection Region")
    plt.fill_between(x, y, 0, where=(x > lower_bound) & (x < upper_bound), color='green', alpha=0.5, label="Fail to Reject Region")

    # Plot the Z-statistic
    plt.axvline(z_statistic, color='black', linestyle='--', label=f"Z-Statistic = {z_statistic:.2f}")

    # Add critical values
    plt.axvline(lower_bound, color='purple', linestyle='--', label=f"Lower Critical Value = {lower_bound:.2f}")
    plt.axvline(upper_bound, color='purple', linestyle='--', label=f"Upper Critical Value = {upper_bound:.2f}")

    plt.title("Two-Tailed Z-Test Decision Region")
    plt.xlabel("Z-Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
if __name__ == "__main__":
    # Simulate sample data
    np.random.seed(42)  # Set seed for reproducibility
    sample_data = np.random.normal(loc=105, scale=10, size=50)  # Mean=105, SD=10, Size=50
    population_mean = 100  # Known population mean
    population_std = 10     # Known population standard deviation
    alpha = 0.05           # Significance level

    # Perform the two-tailed Z-test
    two_tailed_z_test(sample_data, population_mean, population_std, alpha)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def visualize_errors(population_mean, population_std, sample_mean, sample_std, sample_size, alpha=0.05):
    """
    Visualize Type 1 and Type 2 errors in hypothesis testing.

    Args:
    - population_mean: Mean of the null hypothesis distribution.
    - population_std: Standard deviation of the null hypothesis distribution.
    - sample_mean: Mean of the alternative hypothesis distribution.
    - sample_std: Standard deviation of the sample.
    - sample_size: Size of the sample.
    - alpha: Significance level (default 0.05).
    """
    # Calculate critical values
    z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed test
    lower_bound = population_mean - z_critical * (population_std / np.sqrt(sample_size))
    upper_bound = population_mean + z_critical * (population_std / np.sqrt(sample_size))

    # Generate x values for plotting
    x = np.linspace(population_mean - 4 * population_std, sample_mean + 4 * sample_std, 1000)

    # Null hypothesis distribution
    null_dist = norm.pdf(x, population_mean, population_std / np.sqrt(sample_size))

    # Alternative hypothesis distribution
    alt_dist = norm.pdf(x, sample_mean, sample_std / np.sqrt(sample_size))

    # Visualization
    plt.figure(figsize=(12, 6))

    # Plot null hypothesis distribution
    plt.plot(x, null_dist, label="Null Hypothesis (H0)", color="blue")

    # Plot alternative hypothesis distribution
    plt.plot(x, alt_dist, label="Alternative Hypothesis (H1)", color="green")

    # Shade Type 1 error region
    plt.fill_between(x, null_dist, 0, where=(x <= lower_bound) | (x >= upper_bound),
                     color='red', alpha=0.5, label="Type 1 Error (α)")

    # Shade Type 2 error region
    plt.fill_between(x, alt_dist, 0, where=(x > lower_bound) & (x < upper_bound),
                     color='orange', alpha=0.5, label="Type 2 Error (β)")

    # Add critical value lines
    plt.axvline(lower_bound, color='purple', linestyle='--', label=f"Lower Critical Value ({lower_bound:.2f})")
    plt.axvline(upper_bound, color='purple', linestyle='--', label=f"Upper Critical Value ({upper_bound:.2f})")

    # Add labels and legend
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
if __name__ == "__main__":
    # Parameters
    population_mean = 100      # Mean under null hypothesis (H0)
    population_std = 15        # Standard deviation under H0
    sample_mean = 105          # Mean under alternative hypothesis (H1)
    sample_std = 15            # Standard deviation under H1
    sample_size = 30           # Sample size
    alpha = 0.05               # Significance level

    # Visualize errors
    visualize_errors(population_mean, population_std, sample_mean, sample_std, sample_size, alpha)
```

6. Write a Python program to perform an independent T-test and interpret the results

```python
import numpy as np
from scipy.stats import ttest_ind

# Function to perform an independent T-test
def independent_t_test(group1, group2, alpha=0.05):
    # Perform the independent T-test
    t_statistic, p_value = ttest_ind(group1, group2)

    # Output results
    print(f"Group 1 Data: {group1}")
    print(f"Group 2 Data: {group2}")
    print(f"T-Statistic: {t_statistic:.2f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpret the results
    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. The means of the two groups are significantly different.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. The means of the two groups are not significantly different.")

# Example usage
if __name__ == "__main__":
    # Simulate sample data for two groups
    np.random.seed(42)  # Set seed for reproducibility
    group1 = np.random.normal(loc=100, scale=10, size=30)  # Mean=100, SD=10, Size=30
    group2 = np.random.normal(loc=105, scale=10, size=30)  # Mean=105, SD=10, Size=30
    alpha = 0.05  # Significance level

    # Perform the independent T-test
    independent_t_test(group1, group2, alpha)
```

7. Perform a paired sample T-test using Python and visualize the comparison results

```python
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Function to perform a paired sample T-test
def paired_t_test(before, after, alpha=0.05):
    # Perform the paired sample T-test
    t_statistic, p_value = ttest_rel(before, after)

    # Output results
    print(f"Before Treatment: {before}")
    print(f"After Treatment: {after}")
    print(f"T-Statistic: {t_statistic:.2f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpret the results
    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}. The treatment has a significant effect.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}. The treatment does not have a significant effect.")

    # Visualize the comparison
    visualize_comparison(before, after)

# Function to visualize the comparison between before and after treatment
def visualize_comparison(before, after):
    x = np.arange(len(before))

    plt.figure(figsize=(10, 6))

    # Plot the before and after treatment values
    plt.plot(x, before, label="Before Treatment", color="blue", marker="o")
    plt.plot(x, after, label="After Treatment", color="green", marker="o")

    # Plot the difference between before and after treatment
    plt.plot(x, after - before, label="Difference", color="red", linestyle="--", marker="x")

    # Adding labels and legend
    plt.title("Paired Sample T-Test Comparison")
    plt.xlabel("Sample Index")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Simulate data for before and after treatment
    np.random.seed(42)  # Set seed for reproducibility
    before_treatment = np.random.normal(loc=100, scale=10, size=30)  # Mean=100, SD=10, Size=30
    after_treatment = before_treatment + np.random.normal(loc=5, scale=5, size=30)  # Increase by mean of 5

    alpha = 0.05  # Significance level

    # Perform the paired sample T-test
    paired_t_test(before_treatment, after_treatment, alpha)
```

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import ttest_1samp, norm
import matplotlib.pyplot as plt

# Function to perform Z-test
def z_test(sample_data, population_mean, population_std, sample_size):
    sample_mean = np.mean(sample_data)
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed p-value
    return z_statistic, p_value

# Function to perform T-test
def t_test(sample_data, population_mean):
    t_statistic, p_value = ttest_1samp(sample_data, population_mean)
    return t_statistic, p_value

# Function to compare Z-test and T-test results
def compare_tests(population_mean, population_std, sample_size, alpha=0.05):
    # Simulate sample data
    np.random.seed(42)  # Set seed for reproducibility
    sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

    # Perform Z-test
    z_stat, z_p_value = z_test(sample_data, population_mean, population_std, sample_size)

    # Perform T-test
    t_stat, t_p_value = t_test(sample_data, population_mean)

    # Output results
    print(f"Z-Test Results:")
    print(f"Z-Statistic: {z_stat:.2f}")
    print(f"P-Value: {z_p_value:.4f}")

    print(f"\nT-Test Results:")
    print(f"T-Statistic: {t_stat:.2f}")
    print(f"P-Value: {t_p_value:.4f}")

    # Interpretation of results
    print("\nInterpretation:")
    if z_p_value < alpha:
        print(f"Z-Test: Reject the null hypothesis (population mean is not equal to sample mean).")
    else:
        print(f"Z-Test: Fail to reject the null hypothesis (population mean is equal to sample mean).")

    if t_p_value < alpha:
        print(f"T-Test: Reject the null hypothesis (population mean is not equal to sample mean).")
    else:
        print(f"T-Test: Fail to reject the null hypothesis (population mean is equal to sample mean).")

    # Visualize the sample data distribution and the critical values for Z and T-tests
    visualize_comparison(sample_data, population_mean, z_stat, t_stat)

# Function to visualize comparison of sample data with critical values for Z and T-tests
def visualize_comparison(sample_data, population_mean, z_stat, t_stat):
    plt.figure(figsize=(10, 6))

    # Plot sample data distribution
    plt.hist(sample_data, bins=20, alpha=0.7, color='skyblue', label="Sample Data")

    # Plot population mean
    plt.axvline(population_mean, color='red', linestyle='--', label=f"Population Mean ({population_mean})")

    # Plot Z-test critical value
    plt.axvline(population_mean + z_stat, color='green', linestyle='--', label=f"Z-Test Critical Value ({z_stat:.2f})")

    # Plot T-test critical value
    plt.axvline(population_mean + t_stat, color='orange', linestyle='--', label=f"T-Test Critical Value ({t_stat:.2f})")

    # Adding labels and legend
    plt.title("Comparison of Z-Test and T-Test Results")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Parameters
    population_mean = 100  # Population mean
    population_std = 15    # Population standard deviation
    sample_size = 30       # Sample size
    alpha = 0.05           # Significance level

    # Compare Z-test and T-test results
    compare_tests(population_mean, population_std, sample_size, alpha)
```

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

# Function to calculate the confidence interval for a sample mean
def confidence_interval(sample_data, confidence_level=0.95):
    # Calculate sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    sample_size = len(sample_data)

    # Calculate the critical value for the given confidence level
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error

# Example usage
if __name__ == "__main__":
    # Simulate sample data
    np.random.seed(42)  # Set seed for reproducibility
    sample_data = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, SD=15, Size=50

    # Calculate the confidence interval for the sample mean
    confidence_level = 0.95
    lower, upper, mean, margin_of_error = confidence_interval(sample_data, confidence_level)

    # Output results
    print(f"Sample Mean: {mean:.2f}")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"{int(confidence_level * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
import numpy as np
import scipy.stats as stats

# Function to calculate the margin of error for a given confidence level
def margin_of_error(sample_data, confidence_level=0.95):
    # Calculate sample mean and sample standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    sample_size = len(sample_data)

    # Calculate the critical value for the given confidence level
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    return margin_of_error, sample_mean

# Example usage
if __name__ == "__main__":
    # Simulate sample data
    np.random.seed(42)  # Set seed for reproducibility
    sample_data = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, SD=15, Size=50

    # Define confidence level
    confidence_level = 0.95

    # Calculate the margin of error
    margin, sample_mean = margin_of_error(sample_data, confidence_level)

    # Output results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Margin of Error for {int(confidence_level * 100)}% Confidence Level: {margin:.2f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
# Bayesian Inference using Bayes' Theorem

def bayesian_inference(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    :param prior: The prior probability of the hypothesis P(H)
    :param likelihood: The likelihood of the evidence given the hypothesis P(E | H)
    :param evidence: The marginal likelihood P(E)
    :return: The posterior probability P(H | E)
    """
    # Apply Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Example usage
if __name__ == "__main__":
    # Define prior, likelihood, and evidence
    prior = 0.1  # Prior probability P(H)
    likelihood = 0.8  # Likelihood P(E | H)
    evidence = 0.5  # Marginal likelihood P(E)

    # Calculate the posterior probability
    posterior = bayesian_inference(prior, likelihood, evidence)

    # Output the result
    print(f"Posterior Probability P(H | E): {posterior:.4f}")
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import scipy.stats as stats

# Sample data: Contingency table of two categorical variables
# Let's assume we have two categorical variables: 'Gender' and 'Preference for Product A or B'

# Contingency table (rows: Gender, columns: Product Preference)
# Columns: [Product A, Product B]
# Rows: [Male, Female]
data = np.array([[30, 10],  # Male: 30 prefer A, 10 prefer B
                 [20, 20]]) # Female: 20 prefer A, 20 prefer B

# Perform Chi-square test for independence
chi2, p_value, dof, expected = stats.chi2_contingency(data)

# Output the results
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation of the result
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("There is a significant association between Gender and Product Preference (Reject H0).")
else:
    print("There is no significant association between Gender and Product Preference (Fail to Reject H0).")
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

```python
import numpy as np
import scipy.stats as stats

# Function to perform Chi-square goodness-of-fit test
def chi_square_goodness_of_fit(observed, expected):
    # Calculate the Chi-square statistic and p-value
    chi2_stat, p_value = stats.chisquare(observed, expected)
    return chi2_stat, p_value

# Example usage
if __name__ == "__main__":
    # Observed data: frequencies of categories in the dataset
    observed = np.array([50, 30, 20])  # Example: 50 A's, 30 B's, 20 C's

    # Expected data: frequencies based on the expected distribution
    # Example: if the expected distribution is equal (i.e., 1/3 for each category)
    total = np.sum(observed)
    expected = np.array([total / 3, total / 3, total / 3])  # Equal expected distribution

    # Perform Chi-square goodness-of-fit test
    chi2_stat, p_value = chi_square_goodness_of_fit(observed, expected)

    # Output the results
    print(f"Chi-square Statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation of the result
    alpha = 0.05  # Significance level (5%)
    if p_value < alpha:
        print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
import scipy.stats as stats

# Function to perform F-test
def f_test(sample1, sample2):
    # Calculate the variances of both samples
    var1 = np.var(sample1, ddof=1)  # Sample variance (ddof=1 for sample variance)
    var2 = np.var(sample2, ddof=1)

    # Calculate the F-statistic (larger variance / smaller variance)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate the p-value from the F-distribution
    p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

    return f_statistic, p_value, var1, var2

# Example usage
if __name__ == "__main__":
    # Generate two random samples (e.g., sample1 and sample2)
    np.random.seed(42)  # For reproducibility
    sample1 = np.random.normal(0, 1, 100)  # Sample 1: mean=0, std=1, size=100
    sample2 = np.random.normal(0, 2, 100)  # Sample 2: mean=0, std=2, size=100

    # Perform the F-test
    f_statistic, p_value, var1, var2 = f_test(sample1, sample2)

    # Output the results
    print(f"Variance of Sample 1: {var1:.4f}")
    print(f"Variance of Sample 2: {var2:.4f}")
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation of the result
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check normality assumption using Shapiro-Wilk test
def check_normality(groups):
    normality_results = {}
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        normality_results[f"Group {i+1}"] = (stat, p_value)
    return normality_results

# Function to check equal variance assumption using Levene's test
def check_equal_variance(groups):
    stat, p_value = stats.levene(*groups)
    return stat, p_value

# Function to perform ANOVA and check assumptions
def check_anova_assumptions(groups):
    # Check normality for each group
    normality_results = check_normality(groups)
    print("Normality Test Results (Shapiro-Wilk):")
    for group, (stat, p_value) in normality_results.items():
        print(f"{group}: Statistic = {stat:.4f}, P-value = {p_value:.4f}")

    # Check equal variance assumption
    stat, p_value = check_equal_variance(groups)
    print(f"\nLevene's Test for Equal Variance: Statistic = {stat:.4f}, P-value = {p_value:.4f}")

    # Check if assumptions are satisfied
    alpha = 0.05
    print("\nAssumptions Check:")
    # Check normality
    normality_passed = all(p_value > alpha for _, (stat, p_value) in normality_results.items())
    # Check equal variance
    equal_variance_passed = p_value > alpha

    if normality_passed and equal_variance_passed:
        print("Assumptions satisfied: Normality and Equal Variance.")
    else:
        if not normality_passed:
            print("Normality assumption is violated.")
        if not equal_variance_passed:
            print("Equal variance assumption is violated.")

# Example usage
if __name__ == "__main__":
    # Simulate data for three groups (e.g., Group A, Group B, Group C)
    np.random.seed(42)  # For reproducibility
    group_a = np.random.normal(50, 10, 30)  # Group A: mean=50, std=10, n=30
    group_b = np.random.normal(55, 12, 30)  # Group B: mean=55, std=12, n=30
    group_c = np.random.normal(60, 15, 30)  # Group C: mean=60, std=15, n=30

    # Check assumptions
    check_anova_assumptions([group_a, group_b, group_c])

    # Visualize the data distributions
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=[group_a, group_b, group_c])
    plt.title("Boxplot of Group Data")
    plt.show()

    # Q-Q plot for normality check
    plt.figure(figsize=(8, 6))
    for i, group in enumerate([group_a, group_b, group_c]):
        stats.probplot(group, dist="norm", plot=plt)
        plt.title(f"Q-Q Plot for Group {i+1}")
        plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(50, 10, 30)  # Sample: mean=50, std=10, n=30

# Population variance (hypothesized value)
sigma_squared_0 = 100  # Hypothesized population variance (std^2 = 10^2 = 100)

# Sample size and sample variance
n = len(sample_data)
sample_variance = np.var(sample_data, ddof=1)

# Calculate the Chi-square test statistic
chi_square_stat = (n - 1) * sample_variance / sigma_squared_0

# Degrees of freedom
df = n - 1

# Significance level (alpha)
alpha = 0.05

# Calculate the critical values for a two-tailed test
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Print the test statistic and critical values
print(f"Chi-square Test Statistic: {chi_square_stat:.4f}")
print(f"Critical Values: Lower = {critical_value_lower:.4f}, Upper = {critical_value_upper:.4f}")

# Hypothesis testing interpretation
if chi_square_stat < critical_value_lower or chi_square_stat > critical_value_upper:
    print("\nThe test statistic falls outside the critical region. We reject the null hypothesis.")
    print("There is a significant difference between the sample variance and the hypothesized population variance.")
else:
    print("\nThe test statistic falls within the critical region. We fail to reject the null hypothesis.")
    print("There is no significant difference between the sample variance and the hypothesized population variance.")

# Visualize the Chi-square distribution with the critical region
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-square Distribution (df={df})', color='b')
plt.fill_between(x, y, where=(x < critical_value_lower) | (x > critical_value_upper), color='red', alpha=0.3, label="Rejection Region")
plt.axvline(chi_square_stat, color='g', linestyle='--', label=f'Test Statistic: {chi_square_stat:.4f}')
plt.axvline(critical_value_lower, color='r', linestyle='--', label=f'Critical Value Lower: {critical_value_lower:.4f}')
plt.axvline(critical_value_upper, color='r', linestyle='--', label=f'Critical Value Upper: {critical_value_upper:.4f}')

plt.title("Chi-square Test for Population Variance")
plt.xlabel("Chi-square Statistic")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groupsD

```python
import numpy as np
import scipy.stats as stats

# Simulated data for two groups
# Group 1: 80 successes out of 200 trials
# Group 2: 90 successes out of 220 trials
x1 = 80  # successes in group 1
n1 = 200  # sample size of group 1

x2 = 90  # successes in group 2
n2 = 220  # sample size of group 2

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion
P = (x1 + x2) / (n1 + n2)

# Z-test statistic calculation
z_statistic = (p1 - p2) / np.sqrt(P * (1 - P) * (1 / n1 + 1 / n2))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print results
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis testing interpretation
alpha = 0.05  # significance level
if p_value < alpha:
    print("\nThe p-value is less than 0.05. We reject the null hypothesis.")
    print("There is a significant difference between the proportions of the two groups.")
else:
    print("\nThe p-value is greater than 0.05. We fail to reject the null hypothesis.")
    print("There is no significant difference between the proportions of the two groups.")
```
