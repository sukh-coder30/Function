2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable

```python
data = np.random.randn(1000)
sns.histplot(data, bins=30, kde = True)
plt.title('Distribution of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [45, 32, 67, 85, 29]
}


df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Values', data=df)
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
# Note: we can also use other Seaborn plots like box plot or violin plot to compare categories with more complex numerical data
```

Q4.  Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data_size = 100

df = pd.DataFrame({
    'Category': np.random.choice(categories, data_size),
    'Value': np.random.randn(data_size) * 10 + 50
})
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')

plt.title('Distribution of Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
np.random.seed(42)
n_samples = 100

feature_1 = np.random.randn(n_samples)
feature_2 = feature_1 * 0.8 + np.random.randn(n_samples) * 0.2
feature_3 = feature_1 * -0.6 + np.random.randn(n_samples) * 0.3
feature_4 = np.random.randn(n_samples)

data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

corr_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
