1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
x = np.linspace(0, 10, 100)
y = np.sin(x)
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y', width=600, height=400)

p.line(x, y, legend_label="y = sin(x)", line_width=2, color="blue")

show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns

```python
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

output_notebook()

np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.rand(100) * 50 + 10
colors = np.random.rand(100) * 100

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))

p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y', width=600, height=600)

p.circle(x='x', y='y', size='sizes', color=color_mapper, source=source, fill_alpha=0.6, line_color=None)

show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits,title="Fruit Counts",toolbar_location=None, tools="")

p.vbar(x='fruits', top='counts', width=0.6, source=source,
       fill_color=factor_cmap('fruits', palette=['#718dbf', '#e84d60', '#ddb7b1', '#8dd3c7'], factors=fruits))

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = 'Fruits'
p.yaxis.axis_label = 'Counts'
show(p)
```

 4.Create a Bokeh histogram to visualize the distribution of the given data.
 data_hist = np.random.randn(1000)

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data", background_fill_color="#f5f5f5")

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black", alpha=0.7)

p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'
p.grid.grid_line_color = "white"

show(p)
```

```python
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None)

color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
